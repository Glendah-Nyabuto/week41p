# **MERGING COLUMNS**

```python
#merging 
df['Date Time'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])
autolib = df.sort_values(by='Date Time')
autolib.head(5)
```

```python
# sorting the dataset ny Punlic name and date
autolib_set = autolib.groupby(['Date Time','Public Name', 'City']).sum().sort_values(['Public Name', 'Date Time'], ascending=[True, True])
autolib_set.reset_index(inplace=True)
```

```python
# Creating boundaries where the pupbic names are diffrent to enable the comparison
autolib_set['Comparison'] = autolib_set['Public Name'].eq(autolib_set['Public Name'].shift())
autolib_set.head(5)
```

```python
# creatind a column to shoe the diffrence in Blue car counter
autolib_set['Blue Car Diff']= autolib_set['Bluecar Counter'].shift(1) - autolib_set['Bluecar Counter']
autolib_set.head(5)
```

```python
# Creating the the column for total utilib counter

autolib_set['Utilib Tot'] = autolib_set['Utilib Counter'] + autolib_set['Utilib 1.4 Counter']

#creatting a column for the total utilib counter per hour
autolib_set['Utilib Tot Count']= autolib_set['Utilib Tot'].shift(1) - autolib_set['Utilib Tot']
autolib_set.head(5)
```

```python
autolib_set.loc[autolib_set['Comparison'] == False, 'Blue Car Diff'] = autolib_set['Bluecar Counter']*-1
autolib_set.loc[autolib_set['Comparison'] == False, 'Utilib Tot Count'] = autolib_set['Utilib Tot Count']*-1
```

# **ANALYSIS**

## Answering questions

```python
# 1. The most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
#Explanation:
# From the Blue car diff column we realize that if it is positive it indicates that cars were picked up(rented)
# Which probaly means the hour with the highest Postive Blues car defrence is the hour which most cars were pickes 

auto_paris = autolib_set.loc[autolib_set['City'] == 'Paris']

print('The most popular hour of the day for picking up a shared electric car (Bluecar) is:')
auto_paris.loc[auto_paris['Blue Car Diff']>0].groupby(['Hour']).sum()['Blue Car Diff'].sort_values(ascending =False).head(1)
```

```python
# 2. Popular hour for returning cars (blue car counter)
#Explanation:
#From the Blue car diff column we realize that if it is negative it indicates that cars were returned
# Which probaly means the hour with the lowest negative Blues car defrence is the hour which most cars were returned

print('The Popular hour for returning cars (blue car counter) is:')
auto_paris[auto_paris['Blue Car Diff']<0].groupby('Hour').sum()['Blue Car Diff'].sort_values(ascending = True).head()
```

```python
# 3. Popular hour for picking cars(Utilib Counter and Utilib 1.4 Counter)
#Explanation:
#From the Utilib Tot Count column we realize that if it is positive it indicates that cars were picked up(rented)
# Which probaly means the hour with the highest Postive Utilib Tot Count is the hour which most cars were pickes 

print('The popular hour for picking cars(Utilib Counter and Utilib 1.4 Counter) is:')
auto_paris[auto_paris['Utilib Tot Count']>0].groupby('Hour').sum()['Utilib Tot Count'].sort_values(ascending =False).head(1)
```

```python
# 4. Popular hour for returning cars (Utilib Counter and Utilib 1.4 Counter)
#Explanation
#From the Blue car diff column we realize that if it is negative it indicates that cars were returned
# Which probaly means the hour with the lowest negative Blues car defrence is the hour which most cars were returned

print('The Popular hour for returning cars (Utilib Counter and Utilib 1.4 Counter) is:')
auto_paris[auto_paris['Utilib Tot Count']>0].groupby('Hour').sum()['Utilib Tot Count'].sort_values(ascending =True).head(1)
```
